Variance Inflation Factor (VIF) is a measure used to detect the presence of multicollinearity in regression analysis. Multicollinearity occurs when two or more predictor variables are highly correlated, which can inflate the variance of the estimated regression coefficients and make the model unreliable.



```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif.tail()
```

### ROC curve

```python
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

ROC curve plots the True Positive Rate (TPR) against the False Positive Rate (FPR) at various threshold settings. 
* $TPR = TP/(TP+FN)$ : y-axis
* $FPR = FP/(FP+TN)$ : x-axis

  Diagonal Line (Baseline): A random classifier would produce a diagonal line from (0,0) to (1,1). This line represents a model with no discrimination capability, i.e., random guessing.
Area Under the Curve (AUC):
The AUC is a single scalar value that summarizes the performance of the classifier across all thresholds.
AUC ranges from 0 to 1.
AUC = 1: Perfect classifier.
AUC = 0.5: No discrimination ability, equivalent to random guessing.
AUC < 0.5: Worse than random guessing (usually indicates a problem with the model or the data).
