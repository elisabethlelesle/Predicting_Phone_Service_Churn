```python
dataset['gender'].replace(to_replace=['Female', 'Male'], value=[0,1], inplace=True)

dataset['Partner'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['Dependents'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['PhoneService'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['MultipleLines'].replace(to_replace=['No phone service', 'No', 'Yes'], value=[-1,0,1], inplace=True)

dataset['InternetService'].replace(to_replace=['DSL', 'Fiber optic'], value=[0,1], inplace=True)

dataset['OnlineSecurity'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['DeviceProtection'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['TechSupport'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['StreamingTV'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['StreamingMovies'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['Contract'].replace(to_replace=['Month-to-month','One year','Two year'], value=[0,1,2], inplace=True)

dataset['PaperlessBilling'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)

dataset['PaymentMethod'].replace(to_replace=['Electronic check', 'Mailed check','Bank transfer (automatic)','Credit card (automatic)'], value=[1,2,3,4], inplace=True)

dataset['Churn'].replace(to_replace=['No', 'Yes'], value=[0,1], inplace=True)


print(dataset)
```
